# Plot graph with MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

- 1.lot the line graph 

```python
plt.plot([1,2,3,4,5],[1,4,9,16,25])
plt.show()
```

- 2. scatter plot

```python
plt.scatter([1,2,3,4,5],[1,4,9,16,25])

plt.show()
```

- 3. scatter plot with numpy array

```python
x= np.random.randn(1000) # 표준정규분포를 따르는 난수 1000개 만듦
y= np.random.randn(1000) # 표준정규분포를 따르는 난수 1000개 만듦

plt.scatter(x, y)
plt.show()
```
