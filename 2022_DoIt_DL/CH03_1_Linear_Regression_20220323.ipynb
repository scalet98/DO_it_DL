```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

print ("==================================================")
print ("|||| Dataset Preperation >> load_diabetes()=======")
print ("==================================================")

diabetes = load_diabetes()
print (">>> diabetes =", diabetes)
```

```python
print ("=============================================")
print ("|||| Dictionary key check ===================")
print ("=============================================")
db_keys = diabetes.keys()
print (">>> db_keys =", db_keys)

print ("\n||| CASE-1 =======================================")
keys_ls = list(db_keys)
print ("\n>>> keys_ls =", keys_ls)
db_tot_df1 = pd.DataFrame((keys_ls), columns = ['Keys'])   
print ("\n>>> db_tot_df1 = \n", db_tot_df1)

print ("\n||| CASE-2 =======================================")
db_ls =[]
db_ls_tot =[]

for key in db_keys:
    db_ls = [key]
    db_ls_tot += [db_ls]
print ("\n>>> db_ls_tot = ", db_ls_tot)                

db_tot_df2 = pd.DataFrame(db_ls_tot, columns = ['Keys'])   
print ("\n>>> db_tot_df2 = \n", db_tot_df2)

print ("=============================================")
print ("|||| Dictionary value check using keys ======")
print ("=============================================")
i = 3
print ("\n>>> keys_ls[i] =", keys_ls[i])
```

```python
print ("=============================================")
print ("|||| DATASET Structure check ================")
print ("=============================================")
print(diabetes.data.shape, diabetes.target.shape)

print ("=============================================")
print ("|||| DATASET Browsing =======================")
print ("=============================================")

print (">>> keys_ls =\n", keys_ls)
print (">>> len(keys_ls) =", len(keys_ls))
print ("\n =======================================")

for key in range(len(keys_ls)):
    print ("\n =======================================")
    print(">>> keys_ls["+str(key)+"] =", keys_ls[key])
    print("\n>>> diabetes[keys_ls["+str(key)+"]] =\n", diabetes[keys_ls[key]])
    print ("\n>>> len(diabetes["+str(key)+"]) =", len(diabetes[keys_ls[key]]))
```

```python
def Visualization(Data_col_no): 

    import pandas as pd
    import numpy as np
    from sklearn.datasets import load_diabetes
    import matplotlib.pyplot as plt

    print ("\n==================================================")
    print ("|||| Dataset Preperation >> load_diabetes()=======")
    print ("==================================================")

    diabetes = load_diabetes()
    print (">>> diabetes =", diabetes)
    print ("=============================================")
    print ("|||| DATASET Preparation for Plottins =======")
    print ("=============================================")

    kk = 0
    kk = Data_col_no 

    print ("\n>>> diabetes['"+keys_ls[kk]+"'].shape =", diabetes[keys_ls[kk]].shape)
    print ("\n>>> diabetes['"+keys_ls[kk]+"'] =\n", diabetes[keys_ls[kk]])
    print ("\n----------------------------------------------------------------------------------------")
    print ("\n>>> len(diabetes'"+keys_ls[kk]+"'[:,2]]) =", len(diabetes[keys_ls[kk]][:,2]))
    print ("\n>>> diabetes['"+keys_ls[kk]+"'[:,2]] =\n", diabetes[keys_ls[kk]][:,2])

    print ("\n==================================================")
    print ("|||| DATASET Visualization with Scatter plot =====")
    print ("==================================================")

    plt.scatter(diabetes.data[:, 2], diabetes.target)

    plt.xlabel('x', fontsize =15)
    plt.ylabel('y', fontsize =15)
    plt.title("Scatter Plot for diabetes['"+keys_ls[(kk)]+"'][:,2]", fontsize =18)
    plt.grid()
    plt.show()
```

```python
print ("\n>>> keys_ls =", keys_ls)
print (">>> len(keys_ls) =", len(keys_ls))
print ("\n>>> diabetes['data'] =\n", diabetes['data'])

Data_col_no = 0

Visualization(Data_col_no)
```
