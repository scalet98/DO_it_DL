```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt

print ("==================================================")
print ("|||| Dataset Preperation >> load_diabetes()=======")
print ("==================================================")

diabetes = load_diabetes()
print (">>> diabetes =", diabetes)
```

```python
plt.scatter(diabetes.data[:, 2], diabetes.target)
```

```python
# DATA PREPERATION 

x = diabetes.data[:, 2]
y = diabetes.target
```

```python
# initializing w, b 

w = 1.0
b = 1.0 
```

```python
# Attaining y_hat using first sample data of training dataset 

y_hat = x[0]*w +b
print(y_hat)
```

```python
# Comparing Target data with predicted data
print (y[0])
```

```python
# Change predicted result by controling w value 

w_inc = w + 0.1 
y_hat_inc = x[0] * w_inc + b 

print (y_hat_inc)
```

```python
# Prediction changes rate check by controling w value 

w_rate = (y_hat_inc-y_hat)/ (w_inc-w)

print (w_rate)
```

```python
# weight value update by change rate 
w_new = w + w_rate 
print (w_new)
```

```python
# intercept update by change rate 

b_inc = b + 0.1
y_hat_inc = x[0] * w +b_inc

print (">>> y_hat_inc =", y_hat_inc)

b_rate = (y_hat_inc - y_hat)/ (b_inc-b)
print ("\n>>> b_rate =", b_rate)
```

```python
# weight and intercept upate by backpropergation 

err = y[0] - y_hat
w_new = w + w_rate * err 
b_new = b + 1 * err 

print (">>> w_new =", w_new)
print (">>> b_new =", b_new)
```

```python
y_hat = x[1] * w_new + b_new 
err = y[1] - y_hat 
w_rate = x[1]
w_new = w_new + w_rate * err 
b_new = b_new + 1 * err 

print (">>> w_new =", w_new)
print (">>> b_new =", b_new)
```

```python
# Repeate for all dataset 

for x_i, y_i in zip(x, y):
    y_hat = x_i * w + b 
    err = y_i - y_hat 
    w_rate = x_i
    w = w + w_rate * err 
    b = b + 1 * err 

print (">>> w =", w)
print (">>> b =", b)
```

```python
# plotting results 

plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)

plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])

plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize =15)

plt.grid()
plt.show()
```

```python
# REPEAT VARIOUS EPOCH 

for i in range(1, 100): 
    for x_i, y_i in zip(x, y):
        y_hat = x_i * w + b 
        err = y_i - y_hat 
        w_rate = x_i
        w = w + w_rate * err 
        b = b + 1 * err 

print (">>> w =", w)
print (">>> b =", b)        
```

```python
# plotting results 

plt.scatter(x, y)
pt1 = (-0.1, -0.1 * w + b)
pt2 = (0.15, 0.15 * w + b)

plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])

plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize =15)

plt.grid()
plt.show()
```

```python
# Predict using model 

x_new = 0.18
y_pred = x_new * w + b 

print (y_pred)
```

```python
# plotting results 

plt.scatter(x, y)
plt.scatter(x_new, y_pred) 

plt.xlabel('x', fontsize =15)
plt.ylabel('y', fontsize =15)

plt.grid()
plt.show()
```
